**Binomial Distribution**

The Q-Q plot below appears to be linear (albeit stepwise), meaning that the mean and standard deviation provide useful information about the variable.

```python
import numpy as np
import scipy as sp
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline

#create and evaluate distribution
binomial = np.random.binomial(10, 0.5, 100)
mean = binomial.mean()
stdev = binomial.std(ddof=1)
print("Mean = ",mean)
print("Standard Deviation =",stdev)

# Plot histogram
plt.subplot(1, 2, 1)
plt.hist(binomial,color='c')
plt.axvline(x=mean,color='red')
plt.axvline(x=mean+stdev,color='k',linestyle = 'dashed')
plt.axvline(x=mean-stdev,color='k',linestyle = 'dashed')
plt.title('Binomial Distribution')


# Plot Q-Q plot
binomial.sort()
norm = np.random.normal(0, 1, 100)
norm.sort()
plt.subplot(1, 2, 2)
plt.plot(norm, binomial, "o")
plt.title('Q-Q Plot: Binomial Distribution')
```

**Rayleigh Distribution**

The Q-Q plot below is not perfectly linear, meaning that the usefulness of the mean and standard deviation is questionable. A more appropriate statistic that could be used to describe the data is the median and mode.

```python
#create and evaluate distribution
rayleigh = np.random.rayleigh(1, 100) 
mean = rayleigh.mean()
stdev = rayleigh.std(ddof=1)
print("Mean = ",mean)
print("Standard Deviation =",stdev)

# Plot histogram
plt.subplot(1, 2, 1)
plt.hist(rayleigh,color='c',bins=10, normed=True)
plt.axvline(x=mean,color='red')
plt.axvline(x=mean+stdev,color='k',linestyle = 'dashed')
plt.axvline(x=mean-stdev,color='k',linestyle = 'dashed')
plt.title('Rayleigh Distribution')

# Plot Q-Q plot
plt.subplot(1, 2, 2)
rayleigh.sort()
norm = np.random.normal(0, 1, 100)
norm.sort()
plt.plot(norm, rayleigh, "o")
plt.title('Q-Q Plot: Rayleigh Distribution')
```

**Weibull Distribution**

The Q-Q plot below appears to be exponential, meaning that the distribution is not normal and therefore the mean and standard deviation do not provide useful information about the variable. A more appropriate statistic that could be used to describe the data is the mode.

```python
#create and evaluate distribution
weibull = np.random.weibull(0.5, 100) 
mean = weibull.mean()
stdev = weibull.std(ddof=1)
print("Mean = ",mean)
print("Standard Deviation =",stdev)

# Plot histogram
plt.subplot(1, 2, 1)
plt.hist(weibull,color='c',bins=10, normed=True)
plt.axvline(x=mean,color='red')
plt.axvline(x=mean+stdev,color='k',linestyle = 'dashed')
plt.axvline(x=mean-stdev,color='k',linestyle = 'dashed')
plt.title('Weibull Distribution')

# Plot Q-Q plot
plt.subplot(1, 2, 2)
weibull.sort()
norm = np.random.normal(0, 1, 100)
norm.sort()
plt.plot(norm, weibull, "o")
plt.title('Q-Q Plot: Weibull Distribution')
```

**Gamma Distribution**

The Q-Q plot below is not perfectly linear, meaning that the usefulness of the mean and standard deviation could be regarded as questionable.

```python
#create and evaluate distribution
gamma = np.random.standard_gamma(2, 100)
mean = gamma.mean()
stdev = gamma.std(ddof=1)
print("Mean = ",mean)
print("Standard Deviation =",stdev)

# Plot histogram
plt.subplot(1, 2, 1)
plt.hist(gamma,color='c',bins=10, normed=True)
plt.axvline(x=mean,color='red')
plt.axvline(x=mean+stdev,color='k',linestyle = 'dashed')
plt.axvline(x=mean-stdev,color='k',linestyle = 'dashed')
plt.title('Gamma Distribution')

# Plot Q-Q plot
plt.subplot(1, 2, 2)
gamma.sort()
norm = np.random.normal(0, 1, 100)
norm.sort()
plt.plot(norm, gamma, "o")
plt.title('Q-Q Plot: Gamma Distribution')
```

**Logseries Distribution**

```python
#create and evaluate distribution
logseries = np.random.logseries(0.99, 100)
mean = logseries.mean()
stdev = logseries.std(ddof=1)
print("Mean = ",mean)
print("Standard Deviation =",stdev)

# Plot histogram
plt.subplot(1, 2, 1)
plt.hist(logseries,color='c',bins=10, normed=True)
plt.axvline(x=mean,color='red')
plt.axvline(x=mean+stdev,color='k',linestyle = 'dashed')
plt.axvline(x=mean-stdev,color='k',linestyle = 'dashed')
plt.title('Logseries Distribution')

# Plot Q-Q plot
plt.subplot(1, 2, 2)
logseries.sort()
norm = np.random.normal(0, 1, 100)
norm.sort()
plt.plot(norm, logseries, "o")
plt.title('Q-Q Plot: Logseries Distribution')
```

**Geometric**

The Q-Q plot below appears to be exponential or parabolic, meaning that the distribution is not normal and therefore the mean and standard deviation do not provide useful information about the variable. A more appropriate statistic that could be used to describe the data is the mode.

```python
#create and evaluate distribution
geometric = np.random.geometric(p=0.35, size=100)
mean = geometric.mean()
stdev = geometric.std(ddof=1)
print("Mean = ",mean)
print("Standard Deviation =",stdev)

# Plot histogram
plt.subplot(1, 2, 1)
plt.hist(geometric,color='c',bins=10, normed=True)
plt.axvline(x=mean,color='red')
plt.axvline(x=mean+stdev,color='k',linestyle = 'dashed')
plt.axvline(x=mean-stdev,color='k',linestyle = 'dashed')
plt.title('Geometric Distribution')

# Plot Q-Q plot
plt.subplot(1, 2, 2)
geometric.sort()
norm = np.random.normal(0, 1, 100)
norm.sort()
slope, intercept, r_value, p_value, std_err = stats.linregress(norm, geometric)
plt.plot(norm, geometric, "o")
plt.title('Q-Q Plot: Geometric Distribution')
```

**Common Characteristics of Descriptive Statistics**

In general, the distributions that could be described by the mean and standard deviation were driven by the quantity *and* magnitudes of points on either side of the mean (and more normal distributions had close to equal number of points on either side). In comparison, the distributions in which the mean and standard deviation were not useful were skewed by the magnitudes of the outliers.

**Sum of Normal Distributions**

The sum of normal distributions appears to have a normal distribution as well, meaning that the mean and standard deviation provide useful information about the variable.

```python
rand1 = np.random.normal(5,0.5,1000)
rand2 = np.random.normal(10,1,1000)
rand_sum = rand1 + rand2

mean = rand_sum.mean()
stdev = rand_sum.std(ddof=1)
print("Mean = ",mean)
print("Standard Deviation =",stdev)

plt.subplot(1, 2, 1)
plt.hist(rand_sum,color='c',bins=10, normed=True)
plt.axvline(x=mean,color='red')
plt.axvline(x=mean+stdev,color='k',linestyle = 'dashed')
plt.axvline(x=mean-stdev,color='k',linestyle = 'dashed')
plt.title('Sum of Random Variables')
plt.show()
```
