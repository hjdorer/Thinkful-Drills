1. Determine the five most common journals and the total articles for each. 
2. Calculate the mean, median, and standard deviation of the open-access cost per article for each journal. 
3. Identify the open access prices paid by subject area.

```python
import pandas as pd
import numpy as np
import re

df = pd.read_csv('WELLCOME_APCspend2013_forThinkful.csv', encoding='latin1')
df.columns = ['PMID/PMCID','Publisher','Journal Title','Article title','Cost']

#Consolidate publishers that appear to be most common
publisher = df['Publisher']
for i in range(len(publisher)):
    if ((publisher[i] == 'Elseveier Science')or
          (publisher[i] == 'ELSEVIER')or
          (publisher[i] == 'Elsevier (Cell Press)')or
          (publisher[i] == 'Elsevier / Cell Science')or
          (publisher[i] == 'Elsevier B.V.')or
          (publisher[i] == 'Elsevier Ltd')or
          (publisher[i] == 'Elsevier/Cell Press')):
        publisher[i] = 'Elsevier'
    elif ((publisher[i] == 'PLOS')or
          (publisher[i] == 'PLoS')or
          (publisher[i] == 'Plos')or
          (publisher[i] == 'PLoS (Public Library of Science)')):
        publisher[i] = 'Public Library of Science'
    elif ((publisher[i] == 'Wiley & Son')or
          (publisher[i] == 'Wiley Online Library')or
          (publisher[i] == 'Wiley Subscription Services')or
          (publisher[i] == 'Wiley Subscription Services  Inc.')or
          (publisher[i] == 'Wiley Subscription Serviices Inc')or
          (publisher[i] == 'John Wiley')or
          (publisher[i] == 'JOHN WILEY & SONS')or
          (publisher[i] == 'John Wiley & Sons')or
          (publisher[i] == 'John Wiley & Sons Inc')or
          (publisher[i] == 'John Wiley & Sons, Inc. ')or
          (publisher[i] == 'John Wiley and Sons')or
          (publisher[i] == 'John Wiley and Sons Ltd')or
          (publisher[i] == 'Wiley-Blackwell')or
          (publisher[i] == 'Wiley-Blackwell, John Wiley & Sons')or
          (publisher[i] == 'Wiley/Blackwell')or
          (publisher[i] == 'Wliey-Blackwell')or
          (publisher[i] == 'Wiley VCH')or
          (publisher[i] == 'Wiley-VCH')or
          (publisher[i] == 'Wiley ')or
          (publisher[i] == 'John Wiley & Sons Ltd')):
        publisher[i] = 'Wiley'
    elif ((publisher[i] == 'OUP')or
          (publisher[i] == 'Oxford Journals')or
          (publisher[i] == 'OXFORD UNIVERSITY PRESS')or
          (publisher[i] == 'Oxford Journals (OUP)')or
          (publisher[i] == 'Oxford Univ Press')or
          (publisher[i] == 'Oxford University Press (OUP)')):
        publisher[i] = 'Oxford University Press'
    elif ((publisher[i] == 'Springer - Verlag GMBH')or
          (publisher[i] == 'Springer Science + Business Media')or
          (publisher[i] == 'Springer Verlag')or
          (publisher[i] == 'Springer-Veriag GmbH ')or
          (publisher[i] == 'Springer-Verlag GmbH')or
          (publisher[i] == 'SPRINGER-VERLAG GMBH')or
          (publisher[i] == 'Springer-Verlag GMBH & Ci')or
          (publisher[i] == 'Springer-Verlag GmbH, Heidelberger Platz 3, D-14197 Berlin')or
          (publisher[i] == 'SPRINGER-VERLAG GMBH')):
        publisher[i] = 'Springer'

print(publisher.value_counts())

#Clean cost data and output mean, median and standard deviation
cost = df['Cost']
df['Cost'] = df['Cost'].map(lambda x: ''.join(list(filter(str.isdigit, str(x)))))
for i in range(len(cost)):
    cost[i] = float(cost[i])/100.00
    if cost[i]>13000:
        cost[i] = np.nan

cost_mean = np.mean(cost)
cost_median = np.median(cost)
cost_std = np.std(cost, ddof=1)
print ("Mean = ",cost_mean,"Median = ",cost_median,"Std Dev = ",cost_std)

#Group journals by subject areas including:
#Chemistry, physics, medicine, psychology and biology
article = df['Article title']
subject = []
for i in range(len(publisher)):
    if ((re.search('ACS' , publisher[i]))or
        (re.search('Chemical' , publisher[i]))):
        subject.append('Chemistry')
    elif ((re.search('Immun' , publisher[i]))or
          (re.search('Genet' , publisher[i]))or
          (re.search('BioMed' , publisher[i]))or
          (re.search('BMC' , publisher[i]))or
          (re.search('BMJ' , publisher[i]))or
          (re.search('Physio' , publisher[i]))or
          (re.search('Endocrine' , publisher[i]))or
          (re.search('Medicin' , publisher[i]))or
          (re.search('Neurosci' , publisher[i]))or
        (re.search('Hematology' , publisher[i]))):
        subject.append('Medicine')
    elif ((re.search('Biolo' , publisher[i]))or
          (re.search('Microbio' , publisher[i]))or
          (re.search('ASBMB' , publisher[i]))or
          (re.search('Biophys' , publisher[i]))or
          (re.search('Biosci' , publisher[i]))or
        (re.search('Biochem' , publisher[i]))):
        subject.append('Biology')
    elif ((re.search('Psycho' , publisher[i]))):
        #(re.search('chemical' , publisher[i]))):
        subject.append('Psychology')
    else:
        subject.append('Undetermined')


#Append subject list to dataframe and output counts and avg cost by subject
df['Subject'] = pd.Series(subject, index=df.index)
df['Subject'].value_counts()

#Getting error with the following code to calculate the mean. Possibly related to the NaN values.
#df.groupby('Subject')['Cost'].mean()
```
