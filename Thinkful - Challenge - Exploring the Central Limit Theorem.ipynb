**Base Code**

```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
%matplotlib inline

pop1 = np.random.binomial(10, 0.2, 10000)
pop2 = np.random.binomial(10,0.5, 10000) 

plt.subplot(1, 2, 1)
plt.hist(pop1, alpha=0.5, label='Population 1') 
plt.hist(pop2, alpha=0.5, label='Population 2') 
plt.legend(loc='upper right')
plt.title('Population Distribution')
plt.show()

sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

plt.subplot(1, 2, 2)
plt.hist(sample1, alpha=0.5, label='sample 1') 
plt.hist(sample2, alpha=0.5, label='sample 2') 
plt.legend(loc='upper right')
plt.title('Sample Distribution')
plt.show()

print("Sample 1 Mean = ",sample1.mean())
print("Sample 2 Mean = ",sample2.mean())
print("Sample 1 Standard Deviation = ",sample1.std())
print("Sample 2 Standard Deviation = ",sample2.std())
```

**Question 2**

Change the population value (p in the NumPy documentation) for pop1 to 0.3, then take new samples and compute the t-statistic and p-value.

```python
from scipy.stats import ttest_ind
pop1 = np.random.binomial(10, 0.3, 10000)
pop2 = np.random.binomial(10,0.5, 10000) 

plt.subplot(1, 2, 1)
plt.hist(pop1, alpha=0.5, label='Population 1') 
plt.hist(pop2, alpha=0.5, label='Population 2') 
plt.legend(loc='upper right')
plt.title('Population Distribution')
plt.show()

sample1 = np.random.choice(pop1, 1000, replace=True)
sample2 = np.random.choice(pop2, 1000, replace=True)

plt.subplot(1, 2, 2)
plt.hist(sample1, alpha=0.5, label='sample 1') 
plt.hist(sample2, alpha=0.5, label='sample 2') 
plt.legend(loc='upper right')
plt.title('Sample Distribution')
plt.show()

print("Sample 1 Mean = ",sample1.mean())
print("Sample 2 Mean = ",sample2.mean())
print("Sample 1 Standard Deviation = ",sample1.std())
print("Sample 2 Standard Deviation = ",sample2.std())
print(ttest_ind(sample2, sample1, equal_var=False))
```
